```python
import numpy as np
import requests
from tqdm import tqdm_notebook
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
```

```python
headers = {'accept': '*/*', 
          'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
```

```python
base_url = 'https://hh.ru/search/vacancy?text=Data+scientist&page='
```

```python
def hh_parse(base_url, headers):
    jobs = []
    for i in tqdm_notebook(range(100)):
        #base_url = 'https://hh.ru/search/vacancy?area=1&clusters=true&enable_snippets=true&page=' + str(i)
        #url = 'https://api.hh.ru/vacancies'
        #url = 'https://hh.ru/search/vacancy?text=Data+scientist&page=' +str(i)
        par = {'text': 'data scientist', 'area':'1','per_page':'10', 'page':i}
        session = requests.Session()
        request = session.get(base_url, headers=headers, params=par)
        if request.status_code == 200:
            soup = bs(request.content, 'html.parser')
            divs = soup.find_all('div', attrs = {'data-qa': 'vacancy-serp__vacancy'})
            for div in divs:
                if div.find('div', attrs = {'data-qa': 'vacancy-serp__vacancy-compensation'}) != None:
                    if div.find('a', attrs = {'data-qa': 'vacancy-serp__vacancy-title'}) != None:
                         if div.find('a', attrs = {'data-qa': 'vacancy-serp__vacancy-employer'}) != None:
                            salary = div.find('div', attrs = {'data-qa': 'vacancy-serp__vacancy-compensation'}).text
                            title = div.find('a', attrs = {'data-qa': 'vacancy-serp__vacancy-title'}).text
                            company = div.find('a', attrs = {'data-qa': 'vacancy-serp__vacancy-employer'}).text
                            #print(salary)
                            jobs.append({
                            'company' : company,
                            'title' : title,
                            'salary' : salary
                            })

        else:
            print('ERROR')

    return jobs
```

```python
jobs = hh_parse(base_url, headers)
```
